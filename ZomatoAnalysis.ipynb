# Identifying and removing missing values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

# Identifying and dropping duplicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

# EDA

```python
df.columns
```

```python
city_count=df.groupby('City')['Restaurant Name'].count().sort_values(ascending=False)
print(city_count.head(10))
```

```python
city_rating=df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
print(city_rating)
```

```python
df['Has Online delivery'].value_counts()
```

```python
deliveryrating=df.groupby('Has Online delivery')['Aggregate rating'].mean()
print(deliveryrating)
```

```python
df['Average Cost for two'].value_counts()
```

```python
costrating=df.groupby('Average Cost for two')['Aggregate rating'].mean()
print(costrating)
```

# Visualization

```python
city_rating.head(10).plot(kind='bar')
plt.title('Average Ratings by City')
plt.show()
```

```python
city_count.head(10).plot(kind='bar')
plt.title('City vs Restaurant count')
plt.xlabel('City')
plt.ylabel('Number of restaurants')
plt.show()
```

```python
deliveryrating.plot(kind='bar')
plt.title('Online Delivery vs Ratings')
plt.xlabel('Online Delivery')
plt.ylabel('Average Rating')
plt.show()
```

```python
sns.scatterplot(x='Average Cost for two',y='Aggregate rating',data=df)
plt.title('Average Cost vs Ratings')
plt.xlabel('Average Cost')
plt.ylabel('Ratings')
plt.show()
```

# Insights

- Most restaurants are concentrated on top cities

- Ratings are mostly between 3.5 to 5

- Online delivery is a key factor for high ratings

- Higher cost restaurants tend to have better ratings
